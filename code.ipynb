```python
#importing
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
data=pd.DataFrame(pd.read_csv("weatherHistory.csv"))
data
```

```python
data=data.drop(['Formatted Date','Visibility (km)','Daily Summary','Loud Cover','Pressure (millibars)',
               'Apparent Temperature (C)','Wind Speed (km/h)', 'Wind Bearing (degrees)'],axis=1)
data.head()
```

```python
l=LabelEncoder()
data['Summary']=l.fit_transform(data['Summary'])
data['Precip Type']=l.fit_transform(data['Precip Type'])
data.head()
```

```python
print("Missing values in data:\n",data.isnull().sum())
```

```python
data.columns
```

```python
x=data[['Precip Type', 'Temperature (C)','Humidity']]
y=data['Summary']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("mse=",mse)
print("mae=",mae)
print("r2=",r2)
```
